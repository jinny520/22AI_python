조건을 만드는 비교 연산자


```
a > b    # a가 b보다 크다면
a < b    # a가 b보다 작다면
a >= b    # a가 b보다 크거나 같다면
a <= b    # a가 b보다 작거나 같다면
a == b    # a와 b가 같다면
a != b    # a와 b가 틀리다면
```

# **사용자 정의 함수**

````
# 간단한 함수

def greeting() :
    print('안녕하세요.')
    print('영화 관람 금액은 10000원입니다.')

greeting()
````

````
# 매개변수가 있는 함수 (어떤 값을 전달하느냐에 따라 결과가 달라지기를 바랄 때)

adult = 10000
student = 8000

def cal_total(x,y) :
    total = adult*x + student*y
    print('지불할 금액은', total, '원입니다.')

cal_total(2,1)
````

````
# 리턴값이 있는 함수 (함수 실행을 통해 얻은 결과값을 활용하고 싶을 때)

def cal_cost(x,y) :
    total = adult*x + student*y
    return total

money = 50000
pay = cal_cost(2,1)*0.8
left = money - pay

print(left)
````

````
# 파이썬에서는 여러 개의 리턴값을 한 번에 받을 수 있음 

def calculate(x,y) :
    result1 = x+y
    result2 = x-y
    return result1, result2

sum, difference = calculate(5, 3)

print(sum, difference)
````

````
# 리스트 append할 때 주의할 점

data = [0, 1, 2, 3]

def change() :    
    data.append(4)
    print(data)

change()
change()
change()
````

````
def change2() :
    while len(data) > 4 :
        data.pop()
    
    data.append(4)
    print(data)

change2()
change2()
change2()
````
