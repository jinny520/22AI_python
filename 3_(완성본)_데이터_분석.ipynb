# 그래프 명령어 정리 
**그래프 그리는 포맷**
```
import matplotlib.pyplot as plt

plt.title('그래프 제목')
# --- 그래프 형태에 따라 알맞은 명령어 넣기 ---
plt.show()
```
    

**그래프 유형별 명령어**
```
# ---- 그래프 유형별 명령어
# 꺾은선그래프
plt.plot([x축 리스트], [y축 리스트])

# 막대그래프 세로
plt.bar([x축 리스트], [y축 리스트])

# 막대그래프 가로
plt.barh([x축 리스트], [y축 리스트])

# 히스토그램 (x축, y축을 구분하지 않고 하나의 리스트만 넣음)
plt.hist( [리스트], bins = 구간개수 )

# 원 그래프 (x축, y축을 구분하지 않고 하나의 리스트만 넣음)
plt.pie( [리스트], labels = [데이터 구간 이름])

# 산점도
plt.scatter([x축 리스트], [y축 리스트])
```

**그래프 명령어 내에 추가 가능한 옵션**

```
color = '색상이름'

# plt.pie(), plt.boxplot()은 labels 사용
labels = [범례 리스트]

# plt.pie()에 퍼센트 자동 표시
autopct = ‘%1.1%%’
```

**기타 사용가능한 명령어**


```
plt.legend()            # 범례 나타내기
```

#### 실습 9 : x축은 국가명, y축은 금메달개수 꺾은선 그래프 그리기 


````
# x축은 국가명, y축은 금메달개수 꺾은선 그래프 그리기 
nation = []
gold = []

for row in olympic : 
    nation.append(row[1])
    gold.append(row[2])

print(nation)
print(gold)
````

#### 인공지능 프로그래밍 - 간단한 순서



1.   인공지능 모델 생성하기
2.   모델에 데이터를 넣고 학습시키기
3.   새로운 데이터를 잘 예측하는지 확인해보기



````
# 1. 필요한 속성만 추출하기 - 꽃잎 길이, 꽃잎 너비
petal_len_wid = []
species = []

for row in data : 
    # iris 엑셀 데이터로 보여주기 -------------
    petal_len_wid.append([row[3], row[4]])
    species.append(row[5])
    
````

````
from sklearn.tree import DecisionTreeClassifier      # 의사결정트리 기능 포함시키기

# 2. 의사결정트리 모델 생성
model = DecisionTreeClassifier(max_depth= 2)         # 의사결정트리의 최대 깊이
````

````
# 3. data의 꽃잎길이와 꽃잎너비로 품종을 판별하도록 학습
model.fit(petal_len_wid, species) 
````

````
# 새로운 데이터가 어떤 품종인지 맞혀보게 하기
# 1) 꽃잎 길이 = 2, 꽃받침 너비 = 0.5
model.predict([[2, 0.5]])
````

````
# 2) 꽃잎 길이 = 5.5, 꽃받침 너비 = 2.0
model.predict([[5.5, 2.0]])
````
