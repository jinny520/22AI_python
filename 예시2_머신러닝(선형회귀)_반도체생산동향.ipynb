## 5. 알고리즘 설계


* 문제 해결을 위한 5단계<br>
##### 1) 데이터를 읽어온다.
##### 2) 꺾은선 그래프와 산점도를 그려본다
##### 3) 2004년부터 최근까지 시스템 반도체 수출량 데이터를 학습한다.
##### 4) 학습한 데이터를 바탕으로 2030년까지 수출량을 예측해본다.
##### 5) 예측한 결과를 시각화해서 나타낸다.

## 6. 코드작성
*   문제 해결을 위한 알고리즘을 코드로 구현






```python


import pandas as pd
df = pd.read_csv('한국의_시스템_반도체_생산동향.csv', encoding='cp949')
df.head()

# 비어있는 데이터(NaN) 처리하기
# df = df.dropna()
df = df.fillna(method='bfill')
# df 데이터의 정보 알아보기
df.info()

# 연도, 평균기온 데이터 저장하기
year = df['연도'].values
export1 = df['시스템 반도체 생산량(억불)'].values

# 꺾은선 그래프 그리기
import matplotlib.pyplot as plt
plt.plot(year, export1)
plt.show()


# 산점도(scatter) 그리기
import matplotlib.pyplot as plt
plt.scatter(year,export1)
plt.show()


# 학습에 필요한 형태로 데이터 모양 가공하기
# numpy : numeric python, 주로 선형대수학 등에 활용
import numpy as np
year = np.reshape(year,(-1,1))
export1 = np.reshape(export1,(-1,1))

# sklearn : scikit-learn, 파이썬에서 머신러닝에 가장 많이 사용되는 라이브러리
from sklearn.linear_model import LinearRegression

# 선형 회귀 모델 학습(fit) - 한 줄로 끝
model = LinearRegression().fit(year, export1)
# 직선의 기울기와 절편 확인
model.coef_, model.intercept_

model2 = LinearRegression().fit(year[:-8], export1[:-8])
model2.coef_, model2.intercept_

model3 = LinearRegression().fit(year[-8:], export1[-8:])
model3.coef_, model3.intercept_

# 전체적으로 예측하기
year2004 = np.arange(2004, 2030).reshape(-1,1)
pred1 = model.predict(year2004)

before2013 = np.arange(2004, 2013).reshape(-1,1)
pred2 = model2.predict(before2013)

after2013 = np.arange(2013, 2030).reshape(-1,1)
pred3 = model3.predict(after2013)

# 원본 그래프와 선형 회귀로 예측한 그래프 각각 그리기
import matplotlib.pyplot as plt
plt.figure(dpi = 200)
plt.style.use('ggplot')
plt.scatter(year[:-11], export1[:-11], color = 'b')
plt.scatter(year[-11:], export1[-11:], color = 'r')
plt.plot(year2004, pred1, 'g', label = '2004~2030')
plt.plot(before2013, pred2, 'b:', label = '2004~2013')
plt.plot(after2013, pred3, 'r--', label = '2013~2030')
plt.legend()
plt.show()
```



*   결과 해석 <br>
model  : 전체(2004~ 2013)데이터로 학습 <br>
model2 : 2004~ 2013 데이터로 학습<br>
model3 : 최근 연도(2013~2021)데이터로 학습 <br>
하여 2030년까지 시스템 반도체 수출량 변화 예측해았다.최근 5년간이 시스템반도체가 수출량이 증가하는 속도가 빠르다는 것을 알 수 있다.
