```python
import csv 
 
f = open('한국 석탄 소비.csv',encoding="cp949")         # 파일 오픈
data = csv.reader(f)         # 파일 읽기 reader
next(data)                   # 파일 헤더 건너뛰기 
data = list(data)            # 파일 리스트로 바꾸기
```

```python
year= []
amount=[]
for row in data:
    year.append(row[0])
    amount.append(int(row[1]))
print(year)
print(amount)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.xlabel('년도')
plt.ylabel('소비량(Mt)')
plt.title("한국 석탄 소비량")
plt.bar(year,amount)
plt.show()
```

### * 최종 질문 작성하기
#### <font color='blue' font size=5> 2015년(파리기후협약)부터 2020(코로나이후)까지의 석탄소비의 변화량과, 2015년부터 2019(코로나이전)까지의 변화량에 큰 차이가 있을까? </font>

## 5. 알고리즘 설계


* 문제 해결을 위한 5단계<br>
##### 1) 2015~2020년까지의 석탄사용량으로 model1을 학습시킨다.
##### 2) 2015~2019년까지(코로나 이전)의 석탄 사용량으로 model2를 학습시킨다.
##### 3) model1으로 2050년까지의 석탄 사용량을 예측한다.
##### 4) model2로 2050년까지의 석탄 사용량을 예측한다.
##### 5) model1의 추정치와 model2의 추정치를 비교한다.

## 6. 코드작성
*   문제 해결을 위한 알고리즘을 코드로 구현






```python
import numpy as np
year = np.reshape (year,(-1,1))
amount = np. reshape(amount, (-1,1))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(year[-5:], amount[-5:])
```

```python
model.coef_,model.intercept_
```

```python
involve2020 = np.arange(2015,2050).reshape(-1,1)
pred1 = model.predict(involve2020)
```

```python
model2=LinearRegression().fit(year[-5:-1],amount[-5:-1])
```

```python
model2.coef_, model2.intercept_
```

```python
exclude2020 = np.arange(2015,2050).reshape(-1,1)
pred2 = model2.predict(exclude2020)
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi = 200)
plt.style.use('ggplot')
plt.plot(involve2020, pred1, 'r', label = '2015~2020')
plt.plot(exclude2020, pred2, 'b:', label = '2015~2019')
plt.legend()
plt.show()
```

*   결과해석 <br>
model : 최근 5년간(코로나 2020 포함) <br>
model2: 최근 5년간(코로나2020 이전) 데이터로 학습시켜 <br>
2015부터 2050년까지 석탄사용량 예측 <br>
코로나 영향일까? 2020년에 석탄 사용량 줄어(빨간선) <br>
